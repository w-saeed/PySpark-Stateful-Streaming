```python
import re

file_path = 'terminal_logs.txt' 

# Open the file in read mode
with open(file_path, 'r') as file:
    lines = file.readlines()

output_lines = [line.strip() for line in lines if '500 milliseconds' in line.lower()]
text_after_period = [sentence.split(',')[-1].strip() for sentence in output_lines]
numbers_from_text = [re.findall(r'\d+', text) for text in text_after_period]

# Flatten the list of lists into a single list of numbers
milliseconds = [int(num) for sublist in numbers_from_text for num in sublist]
milliseconds = milliseconds[:3500]


batch = list(range(1, 3501))
```

Diagram

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))

plt.plot(batch, milliseconds, label='Linear Function')

# Add labels and title
plt.xlabel('Batches')
plt.ylabel('Time in milliseconds')
plt.title('Processing Time per Batch in Milliseconds')

# Show the plot
plt.show()
```
